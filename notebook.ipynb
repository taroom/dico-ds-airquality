### **Explore Data Set**

```python
#polutan
polutan_list = ["PM2_5", "PM10", "SO2", "NO2", "CO", "O3"]
yearly_avg = airquality_df.groupby("year")[polutan_list].mean(); 
```

```python
correlations = airquality_df[['RAIN'] + polutan_list].corr()

# Menampilkan  korelasi antara curah hujan dan setiap polutan
print(correlations['RAIN'][polutan_list])
# terlihat curah hujan tidak terlalu berpengaruh ke polutan
```

## Visualization & Explanatory Analysis

```python
plt.figure(figsize=(12, 8))

# Buat plot untuk setiap polutan utama
for polutan in polutan_list:
    plt.plot(yearly_avg.index, yearly_avg[polutan], marker='o', label=polutan)

# Tambahkan judul, label, dan legenda
plt.title('Tren Rata-rata Kualitas Udara dari Tahun ke Tahun')
plt.xlabel('Tahun')
plt.ylabel('Rata-rata Konsentrasi Polutan')
plt.legend(title='Polutan')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))

for i, polutan in enumerate(polutan_list, 1):
    plt.subplot(2, 3, i)
    sns.scatterplot(data=airquality_df, x='RAIN', y=polutan)
    plt.title(f'Curah Hujan vs {polutan}')
    plt.xlabel('Curah Hujan (mm)')
    plt.ylabel(f'{polutan} (µg/m³)')

plt.tight_layout()
plt.show()
```
